```python
import pandas as pd
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.impute import MissingIndicator,SimpleImputer
```

```python
df = pd.read_csv("Datasets/Titanic-Dataset.csv",usecols = ['Age',"Fare","Survived"])
df.head()
```

```python

X = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X_train.head()
```

# Without using MissingIndicator

```python
si = SimpleImputer()
X_train_trf = si.fit_transform(X_train)
X_test_trf = si.transform(X_test)
```

```python
X_train_trf
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(X_train_trf,y_train)
y_pred = clf.predict(X_test_trf)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

# Using MissingIndicator

```python
mi = MissingIndicator()
mi.fit(X_train)
```

```python
mi.features_ # It represent that in 1st column null values are there
```

```python
x_train_missing = mi.transform(X_train)
```

```python
x_train_missing
```

```python
X_train["Age_NA"] = x_train_missing
X_train.head()
```

```python
x_test_missing = mi.transform(X_test)
X_test["Age_NA"] = x_test_missing
X_test.head()
```

```python
si = SimpleImputer()
X_train_trf2 = si.fit_transform(X_train)
X_test_trf2 = si.transform(X_test)
```

```python
clf = LogisticRegression()
clf.fit(X_train_trf2,y_train)
y_pred = clf.predict(X_test_trf2)
accuracy_score(y_test,y_pred)
```

#### It really improves the accuracy

# Automatic Missing Value Indicator with add_indicator

```python
si = SimpleImputer(add_indicator= True)
```

```python
X_train_imputed = si.fit_transform(X_train)
```

```python
X_test_imputed = si.transform(X_test)
```

```python
clf = LogisticRegression()
clf.fit(X_train_imputed,y_train)
y_pred2 = clf.predict(X_test_imputed)
accuracy_score(y_test,y_pred2)
```
