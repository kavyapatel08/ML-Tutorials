### Opening csv file from an URL

```python
import pandas as pd
```

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)
```

```python
# 1) sep parameter --> for example tab se seperate data ko load kar sakte hai
# 2) index_col --> kisi bhi column ko agar index banana hai to
# 3)* names = ['col1','col2'] --> columns k name dene k liye if vo 1st row ko hi header ki tarah treat kar raha hai ho
# 4)* header = 1 --> if apka header row ban gaya hai or headers me unnamed aa raha hai to ye use karke 1st row ko as a header rakh sakte hai
# 5) use_cols = ['col1','col4','col5'] --> if hum data ko load karte time hi kuch specific data columns hi chahte hai to vo likh sakte hai
# 6) squeeze = True --> column ko serires ki tarah treat karega
# 7) skiprows = [0,2] --> hum specific rows skip kar sakte hai / isme hum lambda function bhi use kar sakte hai to filter entire data
# 8) nrows = 100 --> agar hum 100 rows ka data hi chahte hai 
# 9) encoding = 'latin-1' --> zomato.csv jese data ko agar hum load karenge to usme "unicodeDecodeError" aayega bcoz vo data utf-8 encoded nahi hota to hum ye use kar sakte hai
# 10) error_bad_lines = False --> kuch data set aese hote jisme pehle 5 entries hoti hai column me phir 6 or 7 or 8 so isko handle karne k liye ye param unn sari rows ko delete kar deta hai (BX-Books.csv)
# 11)* dtype = {'target':int} --> target column ko int me convert karega for memory utilization
# 12)* parse_dates = ['date'] --> convert date column str to date format or you can combine diff month and year column as newdate => parse_dates = {'newdate':[1,3]} #index of that cols
# 13) converters = {'target_col':func_name} --> jo bhi aapki target column hai usko uss function ke according convert kar dega
# 14)* na_values = ['-','Not Available'] --> in sari entries ko vo as a NaN values treat karega
# 15)* chunksize = 5000 --> data ko equally chunks me divide kar dega for example => df = pd.read_csv("aug_train.csv",chunksize = 5000)
#                                                                                    for i in df:
#                                                                                        print(i.shape)
#                                                                                    output => 194 rows => (50, 2)  (50, 2)  (50, 2)  (44, 2)
                                                                                                            
```

```python
dfs = pd.read_csv(data,chunksize = 50)
for i in dfs:
    print(i.shape)
```
