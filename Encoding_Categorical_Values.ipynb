```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Datasets/customer.csv")
```

```python
data .head()
```

```python
x = data.iloc[:,2:4]
y = data.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 0)
print(x_train.shape,x_test.shape)
```

## Ordinal Encoding on column Review and Education

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories = [["Poor","Average","Good"],["School","UG","PG"]])
oe.fit(x_train)
```

```python
x_train_trans = oe.transform(x_train)
x_test_trans = oe.transform(x_test)
```

```python
x_train_trans = pd.DataFrame(x_train_trans,columns = x_train.columns)
x_test_trans = pd.DataFrame(x_test_trans,columns = x_test.columns)
```

```python
x_train_trans
```

```python
x_test_trans
```

```python
oe.categories_
```

## One Hot Encoding

```python
df = pd.read_csv("Datasets/cars.csv")
```

```python
df.head()
```

### Using Pandas get_dummies() method

```python
pd.get_dummies(df,columns = ["fuel","owner"],dtype = int)
```

### K-1 One hot Encoding

```python
pd.get_dummies(df,columns = ["fuel","owner"],dtype = int,drop_first = True)
```

#### “The problem with get_dummies() is that it doesn’t store the mapping or order of categories. This can lead to inconsistent columns between training and testing data. That’s why in production we often use sklearn.preprocessing.OneHotEncoder instead, which remembers the categories.”

### One Hot Encoding using sklearn

```python
df['fuel'].value_counts()
```

```python
df['owner'].nunique()
```

```python
x = df.iloc[:,0:4]
y = df.iloc[:,-1]
print(x.shape,y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 0)
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop = 'first',dtype = 'int32', sparse_output=False)
```

```python
x_train_trans = ohe.fit_transform(x_train[['fuel','owner']])
x_test_trans = ohe.transform(x_test[['fuel','owner']])
```

```python
x_train_trans.shape
```

```python
x_train_trans
```

```python
new_data = np.hstack((x_train[['brand','km_driven']].values,x_train_trans)) # Join 2 arrays horizontally 
```

```python
new_data = pd.DataFrame(new_data)
new_data
```

### One hot encoding on Brand Column (32 Categories)

```python
counts = df['brand'].value_counts()
counts
```

```python
threshold = 100 
```

```python
repl = counts[counts <= threshold].index
repl
```

```python
pd.get_dummies(df['brand'].replace(repl,"Uncommon"),dtype = int).sample(5)
```
